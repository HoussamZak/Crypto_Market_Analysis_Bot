```python
def getprices(exchange, symbol):
    inst = getattr(ccxt, exchange)()
    df = pd.DataFrame(inst.fetchOHLCV(symbol, limit = 60))
    df.columns = ['Time', 'Open', 'High', 'Low', 'Close', 'Volume']
    df.set_index('Time', inplace = True)
    df.index = pd.to_datetime(df.index, unit = 'ms')

    # Cast all as values as float
    df = df.astype(float)
    return df
```

```python
getprices('binance', 'ETH/BTC')
getprices('coinbasepro', 'ETH/USDT')
```

```python
# Iterate over exchanges and call our getprices fct for a specific symbol
# let's store in two lists prices [] and fails for failed data fetching

prices = []
fails = []

for exchange in ccxt.exchanges:
    try:
        prices.append(getprices(exchange, 'BTC/USDT'))
        print('prices successfully pulled for ' + exchange)
    except:
        fails.append(exchange)
        print('Price pulling failed for ' + exchange)

```
